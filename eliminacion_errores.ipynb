```python
import pandas as pd
import numpy as np
```

```python
acceso_internet_tecnologia_provincia = pd.read_csv('Datasets/Acceso a Internet fijo por tecnología y provincia.csv')
acceso_internet_tecnologia_provincia 
```

```python
acceso_internet_tecnologia_provincia.drop(792, inplace=True)
acceso_internet_tecnologia_provincia
```

```python
acceso_internet_tecnologia_provincia['Año'].value_counts()
```

```python
acceso_internet_tecnologia_provincia['Año'] = acceso_internet_tecnologia_provincia['Año'].replace('2019 *', '2019')
acceso_internet_tecnologia_provincia['Año'] = acceso_internet_tecnologia_provincia['Año'].replace('*', np.nan)
acceso_internet_tecnologia_provincia['Año'].value_counts()
```

```python
acceso_internet_tecnologia_provincia['Trimestre'].value_counts()
```

```python
acceso_internet_tecnologia_provincia['Trimestre'] = acceso_internet_tecnologia_provincia['Trimestre'].replace('3 *', '3')
acceso_internet_tecnologia_provincia['Trimestre'] = acceso_internet_tecnologia_provincia['Trimestre'].replace('2 *', '3')
acceso_internet_tecnologia_provincia['Trimestre'] = acceso_internet_tecnologia_provincia['Trimestre'].replace('1 *', '3')
acceso_internet_tecnologia_provincia['Trimestre'].value_counts()
```

```python
acceso_internet_tecnologia_provincia['Provincia'].value_counts()
```

```python
acceso_internet_tecnologia_provincia.dtypes
```

```python
acceso_internet_tecnologia_provincia['ADSL'].value_counts()
```

```python
acceso_internet_tecnologia_provincia.sort_values(by=['ADSL'], ascending=False)
```

```python
acceso_internet_tecnologia_provincia['Cablemodem'].value_counts()
```

```python
acceso_internet_tecnologia_provincia.sort_values(by=['Cablemodem'], ascending=True)
```

```python
acceso_internet_tecnologia_provincia['Cablemodem'] = acceso_internet_tecnologia_provincia['Cablemodem'].replace('- 0', '0')
acceso_internet_tecnologia_provincia['Cablemodem'].value_counts()
```

```python
acceso_internet_tecnologia_provincia.sort_values(by=['Cablemodem'], ascending=True)
```

```python
acceso_internet_tecnologia_provincia['Fibra óptica']
```

```python
acceso_internet_tecnologia_provincia.sort_values(by=['Fibra óptica'], ascending=True)
```

```python
acceso_internet_tecnologia_provincia['Fibra óptica'] = acceso_internet_tecnologia_provincia['Fibra óptica'].replace('- 0', '0')
acceso_internet_tecnologia_provincia['Wireless'] = acceso_internet_tecnologia_provincia['Wireless'].replace('- 0', '0')
acceso_internet_tecnologia_provincia.sort_values(by=['Fibra óptica'], ascending=True)
```
